```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
# Create a subplot with candlestick chart
fig = make_subplots(rows=1, cols=1, shared_xaxes=True)

# Add candlestick trace to the subplot
candlestick_trace = go.Candlestick(x=df['Date'],
                                   open=df['Open'],
                                   high=df['High'],
                                   low=df['Low'],
                                   close=df['Close'],
                                   name='Candlesticks')

# Add the trace to the subplot
fig.add_trace(candlestick_trace)

# Update layout to add x-axis and y-axis titles
fig.update_layout(xaxis_title='Date',
                  yaxis_title='Stock Price',
                  title='Interactive Candlestick Chart')

# Show the interactive graph
fig.show()
```

```python
import dash
from dash import dcc, html
# Assuming 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the frequency to '3M' (3 Months) to group the data by quarters
df['Quarter'] = df['Date'].dt.to_period("4M")

# Find the highest and lowest prices for each quarter
quarterly_highs = df.groupby('Quarter')['High'].max().reset_index()
quarterly_lows = df.groupby('Quarter')['Low'].min().reset_index()

# Lists to store support and resistance levels
support_levels = []
resistance_levels = []

# Create a subplot with candlestick chart
fig = make_subplots(rows=1, cols=1, shared_xaxes=True)

# Add candlestick trace to the subplot
candlestick_trace = go.Candlestick(x=df['Date'],
                                   open=df['Open'],
                                   high=df['High'],
                                   low=df['Low'],
                                   close=df['Close'],
                                   name='Candlesticks')

# Add the trace to the subplot
fig.add_trace(candlestick_trace)

# Plot quarterly support and resistance lines
for i in range(len(quarterly_highs)):
    quarter_high = quarterly_highs.iloc[i]
    quarter_low = quarterly_lows.iloc[i]

    # Store support and resistance levels along with their starting and ending dates
    support_levels.append({
        'Level': quarter_low['Low'],
        'Start Date': quarter_low['Quarter'].start_time,
        'End Date': quarter_low['Quarter'].end_time
    })

    resistance_levels.append({
        'Level': quarter_high['High'],
        'Start Date': quarter_high['Quarter'].start_time,
        'End Date': quarter_high['Quarter'].end_time
    })

    fig.add_shape(type='line',
                  x0=quarter_high['Quarter'].start_time,
                  x1=quarter_high['Quarter'].end_time,
                  y0=quarter_high['High'],
                  y1=quarter_high['High'],
                  line=dict(color='red', width=2, dash='dash'),
                  xref='x',
                  yref='y')

    fig.add_shape(type='line',
                  x0=quarter_low['Quarter'].start_time,
                  x1=quarter_low['Quarter'].end_time,
                  y0=quarter_low['Low'],
                  y1=quarter_low['Low'],
                  line=dict(color='green', width=2, dash='dash'),
                  xref='x',
                  yref='y')

# Print support levels
print("Support Levels:")
for level in support_levels:
    print(f"Level: {level['Level']}, Start Date: {level['Start Date']}, End Date: {level['End Date']}")

# Print resistance levels
print("\nResistance Levels:")
for level in resistance_levels:
    print(f"Level: {level['Level']}, Start Date: {level['Start Date']}, End Date: {level['End Date']}")

# Update layout to add x-axis and y-axis titles
fig.update_layout(xaxis_title='Date',
                  yaxis_title='Stock Price',
                  title='Interactive Candlestick Chart with Quarterly Support and Resistance (2 Months per Quarter)')

# Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div(children=[
    dcc.Graph(
        id='candlestick-chart',
        figure=fig,
        config={'staticPlot': False, 'displayModeBar': True}
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
